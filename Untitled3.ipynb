```python
import pandas as pd

df = pd.read_csv('twitter_training.csv', encoding='latin1', header=None)
df.columns = ['ID', 'Entity', 'Sentiment', 'Tweet']
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='Sentiment', data=df, palette='Set2')
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Tweet Count")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Remove NaNs from 'Tweet' column
df['Tweet'] = df['Tweet'].astype(str)  # Ensures all values are strings

# Combine tweets by sentiment
positive_text = " ".join(df[df['Sentiment'] == 'Positive']['Tweet'])
negative_text = " ".join(df[df['Sentiment'] == 'Negative']['Tweet'])

# Create WordClouds
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

ax[0].imshow(WordCloud(width=600, height=400, background_color='white').generate(positive_text))
ax[0].set_title('Positive Tweets')
ax[0].axis('off')

ax[1].imshow(WordCloud(width=600, height=400, background_color='white').generate(negative_text))
ax[1].set_title('Negative Tweets')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
top_entities = df['Entity'].value_counts().nlargest(5).index
filtered = df[df['Entity'].isin(top_entities)]

plt.figure(figsize=(10,6))
sns.countplot(data=filtered, x='Entity', hue='Sentiment', palette='pastel')
plt.title("Sentiment by Entity")
plt.ylabel("Number of Tweets")
plt.xticks(rotation=45)
plt.legend(title="Sentiment")
plt.show()
```
